## Clip implementation

CLIP (Contrastive Language-Image Pre-training) model, developed bby OpenAi, is a multi-model vision and language model. It maps images and text descriptions to the same latent space, allowing it to determine whether an image and description match. Clip was trained in a contrastive way to predict which captions correspons to which image and description match.

Clip model is trained baed on the Image and Text Encoder. Image Encoder -> ResNet or ViT, Text Encoder -> transformer. During training, the Image and Text Encoder is trained jointly to maximize the cosine similarity between the image and text embedding of real pairs in the batch. while minimizing the cosine similarity between the embedding of incorrect pairs in the batch. The optimization is performed using a symmentric cross entropy loss over these similarity scores.
